# 靜態語言&動態語言 說明
>## 具體來說，<font color =red>靜態語言</font>是指變數是否帶有資料型態，反之則為<font color =green>動態語言</font>。 
>>### - <font color =red>靜態語言</font>寫程序時要聲明所有變量的數據類型
>>### - <font color =green>動態語言</font>編程不用給任何變量指定數據類型。
>>### - 省略變數類型宣告動作 &rarr; 動態語言
